# Simple public/private-key style toy encryption demo
Encrypt: c = m^p mod n

(it focuses on ENCRYPTION + showing the steps)

```python
# imports
from typing import List, Tuple
```

```python
# functions
def char_to_num(ch: str) -> int:
    """Map a/A=0, b/B=1, ..., z/Z=25. Raises if not a letter."""
    if not ch.isalpha():
        raise ValueError(f"Unsupported character '{ch}'. Use letters a-z/A-Z only.")
    return ord(ch.lower()) - ord('a')

def encrypt_number(m: int, p: int, n: int) -> int:
    """Efficient modular exponentiation."""
    return pow(m, p, n)

def encrypt_string_verbose(text: str, p: int, n: int) -> List[int]:
    encrypted: List[int] = []

    print(f"Parameters: p={p}, modulus={n}")
    print(f"Plaintext: {text}")
    print("-" * 60)

    # 1) Convert letters -> numbers
    conversions: List[Tuple[str, int]] = []
    for ch in text:
        if ch.isspace():
            # If you want to support spaces, either skip or handle separately.
            # Here we skip spaces.
            continue
        m = char_to_num(ch)
        conversions.append((ch, m))

    print("Step 1) Letter -> number mapping (a=0 ... z=25):")
    for ch, m in conversions:
        print(f"  {ch} -> {m}")

    print("-" * 60)
    print("Step 2) Encrypt each number: c = m^p mod n")
    for ch, m in conversions:
        c = encrypt_number(m, p, n)
        encrypted.append(c)
        # Show the exact step the way you asked
        print(f"  {ch}: {m}^{p} mod {n} = {pow(m, p)} mod {n} = {c}")

    print("-" * 60)
    print("Final encrypted sequence:")
    print("  " + " ".join(str(x) for x in encrypted))

    return encrypted
```

# Simple public/private-key style toy decryption demo
Decrypt: m = c^s mod n

Shows each ciphertext number being decrypted and then converted back to a letter.

```python
from typing import List
```

```python
def num_to_char(m: int) -> str:
    """Map 0->a, 1->b, ..., 25->z. If outside 0..25, return '?' to show it's invalid."""
    if 0 <= m <= 25:
        return chr(m + ord('a'))
    return "?"  # indicates it doesn't map to a-z under the given encoding

def decrypt_number(c: int, s: int, n: int) -> int:
    """Efficient modular exponentiation."""
    return pow(c, s, n)

def decrypt_sequence_verbose(cipher_nums: List[int], s: int, n: int) -> str:
    print(f"Parameters: s={s}, modulus={n}")
    print(f"Ciphertext numbers: {' '.join(map(str, cipher_nums))}")
    print("-" * 60)

    plaintext_chars: List[str] = []

    print("Step 1) Decrypt each number: m = c^s mod n")
    for c in cipher_nums:
        m = decrypt_number(c, s, n)
        # Show the exact step the way you asked
        print(f"  {c}^{s} mod {n} = {pow(c, s)} mod {n} = {m}")

        ch = num_to_char(m)
        plaintext_chars.append(ch)

    print("-" * 60)
    print("Step 2) Number -> letter mapping (0=a ... 25=z):")
    for c, ch in zip(cipher_nums, plaintext_chars):
        m = decrypt_number(c, s, n)
        if ch == "?":
            print(f"  {m} -> ?  (does not map to a-z; outside 0..25)")
        else:
            print(f"  {m} -> {ch}")

    result = "".join(plaintext_chars)
    print("-" * 60)
    print(f"Final decrypted text: {result}")
    return result
```

```python
import math
import string

# ---------- SETTINGS ----------
p = 7
s = 3
MOD_N = 33

MOD_HASH_OUT = 32
MOD_SUM = 50

# ---------- HELPERS ----------
def char_to_num(ch: str) -> int:
    if not ch.isalpha():
        raise ValueError(f"Only letters allowed. Found: {ch}")
    return ord(ch.lower()) - ord('a')

def num_to_char(x: int) -> str:
    if 0 <= x <= 25:
        return string.ascii_lowercase[x]
    return "?"

def h_single(x: int) -> int:
    # h(x) = floor(e^x + 100) mod 32
    return int(math.floor(math.e**x + 100)) % MOD_HASH_OUT

def list_hash(nums):
    total = sum(nums)
    x = total % MOD_SUM
    hx = h_single(x)
    return total, x, hx

def sign_hash(hx: int) -> int:
    # sig = hx^s mod 33
    return pow(hx, s, MOD_N)
```

```python
# ---------- PART (c) ----------
plaintext = "SaveMe"

print(f"Plaintext: {plaintext}")
print("-" * 60)

nums = [char_to_num(ch) for ch in plaintext]
print("Step 1) Letter -> number (a=0 ... z=25):")
for ch in plaintext:
    print(f"  {ch} -> {char_to_num(ch)}")

print("-" * 60)
total, x, hx = list_hash(nums)
print("Step 2) List-hash input:")
print(f"  sum(nums) = {total}")
print(f"  x = sum(nums) mod 50 = {total} mod 50 = {x}")

print("-" * 60)
print("Step 3) Hash computation:")
print(f"  h({x}) = floor(e^{x} + 100) mod 32 = {hx}")

print("-" * 60)
sig = sign_hash(hx)
print("Step 4) Signature (private-key encryption of hash):")
print(f"  sig = {hx}^{s} mod {MOD_N} = {pow(hx, s)} mod {MOD_N} = {sig}")

final_message = nums + [sig]
print("-" * 60)
print("Final transmitted message (m1 m2 ... mn sig):")
print("  " + " ".join(map(str, final_message)))
```

```python
import math
import string

# ---------- SETTINGS ----------
p = 7
s = 3
MOD_N = 33

MOD_HASH_OUT = 32
MOD_SUM = 50

# ---------- HELPERS ----------
def num_to_char(x: int) -> str:
    if 0 <= x <= 25:
        return string.ascii_lowercase[x]
    return "?"

def h_single(x: int) -> int:
    return int(math.floor(math.e**x + 100)) % MOD_HASH_OUT

def list_hash(nums):
    total = sum(nums)
    x = total % MOD_SUM
    hx = h_single(x)
    return total, x, hx

def sign_hash(hx: int) -> int:
    return pow(hx, s, MOD_N)

def verify_sig(sig: int, hx: int) -> bool:
    # Verification using public key: sig^p mod 33 should equal hx (since hx is small)
    return pow(sig, p, MOD_N) == hx
```

```python
# ---------- PART (d) ----------
received = [12, 20, 18, 19, 0, 5, 0, 17, 14, 2, 10, 18, 1, 0, 2, 7, 4, 18, 7, 14, 2, 10, 18, 5]

msg_nums = received[:-1]
sig_recv = received[-1]

print("Received message:")
print("  " + " ".join(map(str, received)))
print("-" * 60)

print("Step 1) Separate message numbers and signature:")
print("  msg_nums =", msg_nums)
print("  sig =", sig_recv)

print("-" * 60)
total, x, hx = list_hash(msg_nums)
print("Step 2) Compute hash of message list:")
print(f"  sum(msg_nums) = {total}")
print(f"  x = {total} mod 50 = {x}")
print(f"  h({x}) = floor(e^{x} + 100) mod 32 = {hx}")

print("-" * 60)
sig_expected = sign_hash(hx)
print("Step 3) Expected signature (using private key rule):")
print(f"  sig_expected = {hx}^{s} mod 33 = {pow(hx, s)} mod 33 = {sig_expected}")

print("-" * 60)
print("Step 4) Compare signatures:")
print(f"  received sig = {sig_recv}")
print(f"  expected sig = {sig_expected}")
print("  MATCH?" , sig_recv == sig_expected)

print("-" * 60)
print("Step 5) Public-key verification check (sig^p mod 33 == hash):")
print(f"  sig^p mod 33 = {sig_recv}^{p} mod 33 = {pow(sig_recv, p)} mod 33 = {pow(sig_recv, p, 33)}")
print(f"  hash (hx)     = {hx}")
print("  VERIFIED?", verify_sig(sig_recv, hx))

print("-" * 60)
decoded = "".join(num_to_char(n) for n in msg_nums)
print("Step 6) Decode message numbers -> letters:")
print("  decoded =", decoded)

if sig_recv == sig_expected and verify_sig(sig_recv, hx):
    print("\nFINAL: This IS from Alice. Message =", decoded)
else:
    print("\nFINAL: This is NOT from Alice under this scheme (signature check fails).")
```
